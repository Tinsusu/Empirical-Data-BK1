# Bar chart

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
%matplotlib inline
%config InlineBackend.figure_format='retina'
print(f'pandas version:{pd.__version__}')
pd.set_option('display.max_rows',10)
```

```python
df=pd.read_csv('pisa.csv')
df
```

```python
df[:5].plot(kind='bar')
```

```python
df.set_index('Country')[:5].plot(kind='bar',title= 'Pisa Score 2019')
```

```python
df.set_index('Country', inplace=True)
```

```python
df[:5].plot(kind='barh',title='PISA score 2019', figsize=(8,4))
```

```python
df[:5]['Mathematics'].plot(kind='barh',title='PISA Maths score 2019', figsize=(8,4))
```

```python
df[:5][['Mathematics','Reading']].plot(kind='barh',title='PISA Maths score 2019', figsize=(8,4))
```

```python
ax=df[:5][['Mathematics','Reading']].plot(kind='barh',title='PISA Maths score 2019', figsize=(8,4),color=['blue','pink'])
```

```python
ax=df[:5][['Mathematics','Reading']].plot(kind='barh',title='PISA Maths score 2019', figsize=(8,4),color=['blue','pink'])
ax.axvline(500,color="0.6", linestyle="--",linewidth=0.5)
ax.set_xlabel('SCORE')
```

```python
df.nlargest(10,'Reading')['Reading'].plot(kind='barh')
```

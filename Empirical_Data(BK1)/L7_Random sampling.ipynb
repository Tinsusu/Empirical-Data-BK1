### 1. Calculate the resulting proportion of heads P(H) for all the cases.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from numpy.random import randint 
```

```python
#Check a.flip one coin
one_coin_toss = randint(2)
one_coin10 = randint(2, size=10) 
one_coin25 = randint(2, size=25) 
one_coin100 = randint(2, size=100) 
one_coin250 = randint(2, size=250) 
one_coin1000 = randint(2, size=1000) 
print('P(H) of one coin 1 time:',one_coin_toss/2)
print('P(H) of one coin 10 times:',one_coin10.sum()/10)
print('P(H) of one coin 25 times:',one_coin25.sum()/25)
print('P(H) of one coin 100 times:',one_coin100.sum()/100)
print('P(H) of one coin 250 time:',one_coin250.sum()/250)
print('P(H) of one coin 1000 time:',one_coin1000.sum()/1000)
```

```python
#Check b. flip 3 coin
t_coin = randint(2, size=(1,3))
t_coin10 = randint(2, size=(10,3)) 
t_coin25 = randint(2, size=(25,3)) 
t_coin100 = randint(2, size=(100,3)) 
t_coin250 = randint(2, size=(250,3)) 
t_coin1000 = randint(2, size=(1000,3))
print('P(H) of three coins 1 time:',t_coin.sum()/3)
print('P(H) of three coins 10 times:',t_coin10.sum()/30)
print('P(H) of three coins 25 times:',t_coin25.sum()/75)
print('P(H) of three coins 100 times:',t_coin100.sum()/300)
print('P(H) of three coins 250 times:',t_coin250.sum()/750)
print('P(H) of three coins 1000 times:',t_coin1000.sum()/3000)
```

```python
#Check c. flip 100 coin
h_coin = randint(2, size=(1,100))
h_coin10 = randint(2, size=(10,100)) 
h_coin25 = randint(2, size=(25,100)) 
h_coin100 = randint(2, size=(100,100)) 
h_coin250 = randint(2, size=(250,100)) 
h_coin1000 = randint(2, size=(1000,100))
print('P(H) of 100 coins 1 time:',h_coin.sum()/100)
print('P(H) of 100 coins 10 time:',h_coin10.sum()/1000)
print('P(H) of 100 coins 25 time:',h_coin25.sum()/2500)
print('P(H) of 100 coins 100 time:',h_coin100.sum()/10000)
print('P(H) of 100 coins 250 time:',h_coin250.sum()/25000)
print('P(H) of 100 coins 1000 time:',h_coin1000.sum()/100000)
```

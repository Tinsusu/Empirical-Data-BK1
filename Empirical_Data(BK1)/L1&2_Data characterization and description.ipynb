```python
%matplotlib notebook
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
#from mpl_toolkits.basemap import Basemap
# Draw world
#world = Basemap(projection='cyl',lon_0=0) 
#world.drawcoastlines()
# Load online data and plot earth quakes
path_to_file = 'https://earthquake.usgs.gov/earthquakes/feed/v1.0/summary/1.0_month.csv '
df = pd. read_csv(path_to_file) 
#x, y = world(list(df.longitude), list(df.latitude))
x, y = df.longitude, df.latitude
radius = np.power(df.mag,1/2)*10
plt.scatter(x=x, y=y, s=radius,c='red',alpha=0.2, linewidths=0) 
plt.title('Earthquakes within the last month (>1 Richter)')
```

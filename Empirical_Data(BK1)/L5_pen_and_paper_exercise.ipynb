```python
def fibonacci_version_2(N):
    result = []
    a, b = 0, 1
    while a < N:
        result.append(a)
        a, b = b, a+b
    return result

fib = fibonacci_version_2(100)
golden_ratio = fib[10]/fib[10-1]
golden_ratio
```

## SIR MODEL

```python
# Import package
import scipy.integrate
import numpy
import matplotlib.pyplot as plt
```

```python
def SIR_model(y,t,beta,gamma):
    S,I,R = y
    dS_dt = -beta*S*I
    dI_dt = beta*S*I - gamma*I
    dR_dt = gamma*I
    return([dS_dt,dI_dt,dR_dt])
```

```python
#Initial condition
S0 = 0.9
I0 = 0.1
R0 = 0.0
beta = 0.35
gamma = 0.1

#time vector
t = numpy.linspace(0,100,10000)
#Result
solution = scipy.integrate.odeint(SIR_model,[S0,I0,R0],t,args=(beta,gamma))
solution = numpy.array(solution)
```

```python
#plot result
plt.figure(figsize=[6,4])
plt.plot(t, solution[:,0], label="S(t)")
plt.plot(t, solution[:,1], label="I(t)")
plt.plot(t, solution[:,2], label="R(t)")
plt.grid()
plt.legend
plt.xlabel("Time")
plt.ylabel("Proportions")
plt.title("SIR model")
plt.show()
```

# Butterfly

```python
import pandas as pd
import numpy
import matplotlib.pyplot as plt
# Set model parameters
beta, gamma = 8/3, 1/6 # Model parameters N = 1000000
x,y,z = 1,1,1 #Stateatdayzero
```

```python
# Make simulation one day at a time
data = []
for day in range(10000):
    x, y, z = x+10*(y-x)*0.01, y+(x*(28-z)-y)*0.01, z+(x*y-beta*z)*0.01 
    data.append([day,x,y,z]) # Store values
```

```python
# Create DataFrame with simulation data
df = pd.DataFrame(
     data,
     columns=['Day', 'x','y','z']
         )
```

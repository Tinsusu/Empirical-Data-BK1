## Exercise 8.5: Frequencies and mean

### The following exercise will train the concept of observations given in terms of frequencies. In a certain population of European starling (DK: stÃ¦r) there are 5000 nests with young. The distribution of brood size (DK: kuld-stÃ¸rrelse) is given in Table 7.2. Suppose that one brood is chosen at random and the Y be the size of the brood.

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

b = pd.read_csv('brood.csv')
b
```

### 1. Find P(Y = 4) (The probability that the selected brood has 4 young).

```python
y = b.No_nest[3]/b.No_nest.sum()
y
```

### 2. Find P(4 <Y <7)

```python
p = b.No_nest[3:7].sum()/b.No_nest.sum()
p
```

### 3. What is the total number of young in the nests?

```python
b['t'] = b.Brood*b.No_nest
b
```

```python
b.t.sum()
```

### 4. What is the average number of young in a nest?

```python
b.No_nest.mean()
```

### 5. Optional*: What is the standard deviation of the size?

```python
b['P(x)']= b.No_nest/b.No_nest.sum()
b
```

```python
b['xP(x)']= b['Brood'] * b['P(x)']
b
```

```python
c = b['xP(x)'].sum()
print("Average Brood:",c)
```

```python
b['x-u_sqr']= (b['Brood']-c)**2
b
```

```python
b['x-u_sqr*P(X)']= b['x-u_sqr']*b['P(x)']
b
```

```python
ST = (b['x-u_sqr*P(X)'].sum())**0.5
print("Standard deviation of size:",ST)
```
